```python
class People_of_School():

  student_list = []
  teacher_list = []
  staff_list = []

  def __init__(self, id_number, blood_group, full_name, school_email):
    self.id_number = id_number
    self.blood_group = blood_group
    self.full_name = full_name
    self.school_email = full_name + '.@SchoolBoard.com'

  def add_to_list(self):
    pass # This method will be overridden in subclasses


class Staff(People_of_School):

  def __init__(self, id_number, blood_group, full_name, school_email, job, department, pay):
    super().__init__(id_number, blood_group, full_name, school_email)
    self.job = job
    self.department = department
    self.pay = pay
    self.add_to_list()

  def add_to_list(self):
      People_of_School.staff_list.append(self)

class Teachers(People_of_School):

    def __init__(self, id_number, blood_group, full_name, school_email, subject, teaching_grade, experience, pay):
      super().__init__(id_number, blood_group, full_name, school_email)
      self.subject = subject
      self.teaching_grade = teaching_grade
      self.experience = experience
      self.pay = pay
      self.add_to_list()

    def add_to_list(self):
        People_of_School.teacher_list.append(self)


class Students(People_of_School):

  def __init__(self, id_number, blood_group, full_name, school_email, grade, sport_playing):
    super().__init__(id_number, blood_group, full_name, school_email)
    self.grade = grade
    self.sport_playing = sport_playing
    self.add_to_list()

  def add_to_list(self):
      People_of_School.student_list.append(self)

class Principal_and_VicePrincipal(People_of_School):

  def __init__(self, id_number, blood_group, full_name, school_email, pay, staff= None):
    super().__init__(id_number, blood_group, full_name, school_email)
    self.pay = pay

    if staff is None:
      self.staff = []
    else:
      self.staff = staff

  def add_staff(self, stf):
    if stf not in self.staff:
      self.staff.append(stf)

  def remove_staff(self, stf):
    if stf in self.staff:
      self.staff.remove(stf)

  def view_staff(self):
    print("\nStaff:\n")
    for stf in self.staff:
      print(f"Staff ID: {stf.id_number}, Name: {stf.full_name}")

  @classmethod
  def view_students(cls):
      print("\nStudents:\n")
      for student in People_of_School.student_list:
          print(f"Student ID: {student.id_number}, Name: {student.full_name}, Grade: {student.grade}")

  @classmethod
  def view_teachers(cls):
      print("\nTeachers:\n")
      for teacher in People_of_School.teacher_list:
          print(f"Teacher ID: {teacher.id_number}, Name: {teacher.full_name}, Subject: {teacher.subject}")


std_1 = Students("657 825 356", "O-Positive", "Joshit Vadela", "Joshitvadela@SchoolBoard.com", "Grade 9", "Soccer")
std_2 = Students("805 922 547", "B-Positive", "Haris Hashmi", "Harishashmi@SchoolBoard.com", "Grade 9", "Basketball")
std_3 = Students("560 867 422", "O-Negative", "Sarim, Siddiqui", "Sarimsiddiqui@SchoolBoard.com", "Grade 9", "Hockey")

tch_1 = Teachers("789 052 300", "A-Positive", "Gaberial Lee", "Gaberiallee@SchoolBoard.com", "Science", "Grade 9", "20 years", 80000)
tch_2 = Teachers("457 833 704", "A-Positive", "Brain Farnsworth", "Brainfarnsworth@SchoolBoard.com", "Math", "Grade 8", "35 years", 90000)

stf_1 = Staff("859 699 321", "A-Positive", "Ayaan Zubari", "Ayaanzubari@SchoolBoard.com", "Watchman", "Ground Department", 450000)
stf_2 = Staff("432 900 567", "B-Positive", "Bilal Kamran", "Bilalkamran@SchoolBoard.com", "Cleaner", "Ground Department", 450000)
stf_3 = Staff("859 699 321", "B-Positive", "Rayan Khan", "Rayankhan@SchoolBoard.com", "Technician", "Electric Department", 550000)

prn_1 = Principal_and_VicePrincipal("680 009 001", "O-Negative", "John Dylem", "Johndylem@SchoolBoard.com", 1200000, [stf_1])
vc_prn_1 = Principal_and_VicePrincipal("970 350 215", "A-Negative", "James Kindly", "Jameskindly@SchoolBoard.com", 100000, [stf_2])

# view students
prn_1.view_students()
vc_prn_1.view_teachers()

prn_1.add_staff(stf_3)
prn_1.remove_staff(stf_1)
prn_1.view_staff()
```
